```
import math
# Helper to get a positive number from user
def get_pos_num(prompt_txt):
    while True:
        try:
            val = float(input(prompt_txt))
            if val <= 0:
                print("Gotta be greater than zero, please.\n") # error msg
                continue
            return val
        except ValueError:
            print("That's not a number. Try again.\n") # error msg
def calculate_dose_total(patient_w, dose_per_kg):
    return patient_w * dose_per_kg
def check_dose_ok(current_dose, max_allowed):
    return current_dose <= max_allowed
def count_pills_needed(total_dose, pill_strength):
    # Using math.ceil to round up for full pills/vials
    return math.ceil(total_dose / pill_strength)
def print_summary_info(final_dose, num_pills):
    print("\n----- Dosage Breakdown -----")
    print(f"Required Amount : {final_dose:.2f} mg")
    print("Safety Check    : GOOD TO GO")
    print(f"Units Required  : {num_pills}")
    print("----------------------------")
def start_dose_calculator():
    print("=" * 40)
    print("      Drug Dosage Calculation Tool")
    print("=" * 40)
    wt = get_pos_num("Patient's weight in kg: ")
    dose_rate = get_pos_num("Dosage (mg/kg): ")
    max_dose_limit = get_pos_num("Max safe dose in mg: ")
    strength_per_unit = get_pos_num("Strength per tablet/vial (mg): ")
    total_dose_calc = calculate_dose_total(wt, dose_rate)
    print(f"\nCalculated Dose: {total_dose_calc:.2f} mg")
    if not check_dose_ok(total_dose_calc, max_dose_limit):
        print("\n!!! DANGER !!!") # warning
        print("The calculated dose is too high. It's over the limit.")
        print("Stopping here. Double-check everything before giving any meds.")
        return
    needed_units = count_pills_needed(total_dose_calc, strength_per_unit)
    print_summary_info(total_dose_calc, needed_units)
if __name__ == "__main__":
    start_dose_calculator()
```
